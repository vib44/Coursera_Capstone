### READ THE HTML FILE FOR THE WIKIPEDIA PAGE AND CLEAN THE DATAFRAME




```python
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0) 
ll=pd.read_csv('https://cocl.us/Geospatial_data')                           #Obtain the longitudes and latitudes from the csv file
dfs[0]['Borough'].replace('Not assigned',np.nan,inplace=True)             #Replacing 'Not Assigned' with NaN
dfs[0]['Neighbourhood'].replace('Not assigned',dfs[0]['Borough'],inplace=True)       #Assign Borough to Neighbourhoods with 'Not Assigned'
df=dfs[0].dropna(subset=['Borough'],axis=0)                     #Drop rows with 'NaN' in Borough column
df.reset_index(drop=True,inplace=True)            #Reset index
df.shape
```

## Merge Neighbourhoods with same Postal Codes

```python
#We convert dataframes into lists so that the manipulkations become easier
n=df['Neighbourhood'].values.tolist()
boroughs=df['Borough'].values.tolist()
pc=df['Postcode'].values.tolist()
pcd=pc[1:]
v=[]      #List to save new Neighbourhood values
b=[]      #List to save Borough values
p=[]      #List to save Postal code values without repitition
c=''      #To save the string to be input in list v
k=''      #To save the extra neigbourhood names that are to be added in the Neighbourhood column of a particular postal code
r=1       
d=1
t=0
z=''      
for P,PC in enumerate(pc):       #loop to compare the postal codes
    if((P+t)<211):      #There are total 211 elements in pc
        c=n[P+t]        #Neighbourhood name
        z=pc[P+t]       #Postal code
        k=''
        d=1
        p.append(z)     #appending postal code to list p
        b.append(boroughs[P+t])        #appending Borough name to list b
        for i,g in enumerate(pcd):          #loop to compare the postal codes
            if(z==g):                               
                if(c!=n[i+r]):               #To make sure we dont compare the same Neighborhood with itself
                    k=k+','+n[i+r]
                    d=d+1
                    t=t+1       
            else:
                break
        pcd=pcd[d:]                #As the comparision is done we keep on decreasing the list to eliminate the elements that have been compared
        r=r+d
        c=c+k               #Final string of Neighbourhood names
        v.append(c)         #Appending the string in list v
    else:
        break            
    
        
    
```

```python
print(len(p),len(b),len(v))         #Check whether all the 3 lists have same elements so that we can convert them into a single dataframe
```

## Here are few functions borrowed from Foursquare API to obtain nearby venues,longitude,latitude and venue categories

```python
# function that extracts the category of the venue
#In the above cell we see the key 'categories' is key to the name of the venue category hence we use row['categories']
#or row['venue.categories'] in case categories key exist inside any dictionary with key as 'venue'
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
```

### A json file is obtained which is converted to pandas dataframe so that working with it is made easier

```python
venues = results['response']['groups'][0]['items'] #obtaining the part that contains information about venues
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]  #adding column names to the dataframe

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)   #obtaining categories for respective venues with the help of get_category_type

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
print('{} venues were returned by Foursquare.'.format(nearby_venues.shape[0]))
nearby_venues.head()
```

## Obtaining nearby venues and venues information for every Neighborhood

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):  #loop to find nearby values and their information for every Neighborhood
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    return(nearby_venues)
#
```

### We use one hot encoding when dealing with categorical values while k-means clustering

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
# Here we're grouping the venue categories with respect to neighbourhoods
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:         #for loop to find the venue and the frequency of occurence of the venue in a particular Neighbourhood
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):                   #function for finding most common venues
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]    #returns the top ten venues with highest frequency of occurrence in a Neighborhood
```

### Building a dataframe with most common venues for each neighbourhood with th help of function defined in above cell

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)   #Putting values to each row
neighborhoods_venues_sorted.head()
```

## CLUSTERING OF NEIGHBOURHOODS

```python
# set number of clusters
kclusters = 5
toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')

toronto_merged.head() # check the last columns!
```

## Importing geopy.geocoders package to obatain latitude and longitudes for visualizing the folium map

```python
!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

## Visualizing how the Neighbourhoods cluster together
### Neighbourhoods are clustered together in the map and one can see the neighbourhood name by just clicking on the circle and a popup would occur that shows the neighbourhood name

```python
# create map
%matplotlib inline
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
